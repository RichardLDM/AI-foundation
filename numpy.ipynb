```python
import numpy as np

arr = np.array([1, 3, 5, 7])
print("numpy array:", arr)
```

```python
arr2 = np.array([[1, 3, 5, 7], [4, 6, 8, 9]])
print("numpy array2:", arr2)

arr2.shape
```

```python
# dtype()
a = np.array([1, 2, 3])
print(a.dtype)

b = np.array([1.2, 3.4, 5.6])
print(b.dtype)
```

```python
# reshape()
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)
print("shaped array:", b)
```

```python
# astype()
a = np.array([1, 2, 3])
b = a.astype(float)
print("float:", b)
```

```python
# np.zeros(), np.ones() => dummy data
z = np.zeros((2, 3))
o = np.ones((2, 3))

print("zeros:", z)
print("ones:", o)
```

```python
# ndim()
a = np.array([1, 2, 3])
b = np.array([[1, 2, 3], [4, 5, 6]])

print("a ndim:", a.ndim)
print("b ndim:", b.ndim)
```

```python
# arange()
a = np.arange(1, 6)
print("arrange:", a)
```

```python
# sum(), mean(), max(), min()

a = np.array([1, 2, 3, 4])

print("sum:", a.sum())
print("mean:", a.mean())
print("max:", a.max())
print("min:", a.min())
```

```python
# flatten()

a = np.array([[1,2], [3, 4]])
print("flatten:", a.flatten())
```

```python
# T (Transpose)

a = np.array([[1, 2], [ 3, 4]])
print("transpose:", a.T)
```

```python
# dot() @
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])

print("dot/multiplication:", a.dot(b))
```

```python
# random.rand()
a = np.random.rand(2, 3) # => 0.0 and 1.0 / included / 1.0 excluded
print("random rand:", a)
```
